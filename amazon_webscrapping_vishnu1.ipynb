```python
import requests

from bs4 import BeautifulSoup

r = requests.get("https://www.amazon.in/s?i=electronics&rh=n%3A1389401031&dc=&qid=1752132912&rnid=1318502031&ref=sr_nr_p_36_0_0&low-price=66800&high-price=")
c = r.content
print(c)
```

```python
soup=BeautifulSoup(c,"html.parser")
print(soup)
```

```python
all=soup.find_all("div",{"role":"listitem"})
print(all)
len(all)
```

```python
h2=all[0].find("h2", {"class": "a-size-base-plus a-spacing-none a-color-base a-text-normal"})
print(h2)
all[0].find_all("div",{"data-cy":"title-recipe"})[0].text
```

```python
for item in all:   #to display all price value in the website
    print(item.find_all("h2", {"class": "a-size-base-plus a-spacing-none a-color-base a-text-normal"})[0].text)
    #print(item.find_all("div",{"class":"a-row"})[0].text)
    print(item.find_all("span",{"class":"a-price-whole"})[0].text)
```

```python
l=[]
for i in all:   #to display all price value in the website
    d={}
    d["mobile_name"]=i.find_all("h2", {"class": "a-size-base-plus a-spacing-none a-color-base a-text-normal"})[0].text
    d["Row_info"]=i.find_all("div",{"class":"a-row"})[0].text
    d["Price"]=i.find_all("span",{"class":"a-price-whole"})[0].text
    l.append(d)
print(l)
```

```python
import pandas
df=pandas.DataFrame(l)
print(df)
df.to_csv("amazon_mobile.csv")
```
